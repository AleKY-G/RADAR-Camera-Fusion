```python
%%thread_cell
from geometry_msgs.msg import Pose, Vector3, 
import rospy as rp
import math
i = 0
rate = rp.Rate(5)
pub = rp.Publisher('/poser', Vector3, queue_size=10)
while True:
    msg = Vector3()
    msg.x = math.sin(i * 0.1)
    msg.y = math.cos(i * 0.1)
    pub.publish(msg)
    i += 1
    rate.sleep()
```
